Streaming & Batch


Streaming

```python
for chunk in model.stream("Write an essay on Dog"):
    print(chunk.text,end="",flush=True)
```

Batch

```python
responses = model.batch(
    [
        "why dogs are not fly",
        "Why parrots don't barked",
        "Why human in not have AI mind"
    ]
)
for response in responses :
    print(response)
```

```python
responses = model.batch(
    [
        "why dogs are not fly",
        "Why parrots don't barked",
        "Why human in not have AI mind",
        "how human body can breath",
        "how many alogorithm can have in machine learning",
        "suggest a name of AI tools"
    ],
    config = {
        'max_concurrency':5,
    }
)
for response in responses :
    print(response)
```
