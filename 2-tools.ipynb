Tools

```python
from langchain.tools import tool
```

```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = "......"
```

```python
model = init_chat_model("groq:llama-3.1-8b-instant")
response = model.invoke("write a joke on machine learning Algorithm")
print(response.content)
```

```python
@tool
def get_weather(location:str)->str:
    """Get the weather at a location"""
    return f"It's sunny in {location}"

model_with_tools = model.bind_tools([get_weather])    
```

```python
response = model_with_tools.invoke("What's the weather in Banglore")
print(response)
```

```python
response.tool_calls
```

Tools Execution Loop

```python
# step1- Model generate tool calls
message = [{"role": "user","content": "What's weather in Banglore?"}]
ai_msg = model_with_tools.invoke(message)
message.append(ai_msg)
```

```python
# step 2- Execute tools and collect results
for tool_call in ai_msg. tool_calls:
    tool_result = get_weather.invoke(tool_call)
    message.append(tool_result)
```

```python
# step 3- pass reult back to model for final response
final_response = model_with_tools.invoke(message)
print(final_response.text)
```
